```python
def parse_value(values):
    each_value = []
    for i in values.strip().split(' '):
        if i == '':
            each_value.append(0.0)
        else:
            each_value.append((i))
    return each_value
```

```python
parse_value(file_content[1])
```

```python
def item_dict(headers, values):
    data_dict = {}
    for header, value in zip(headers, values):
        data_dict[header] = value
    return data_dict
    
```

```python
header = parse_headers(file_content[0])
value = parse_value(file_content[1])
item_dict(header, value)
```

```python
def read_csv(path):
    result = []
    with open(path, 'r') as csv_file:
        lines = csv_file.readlines()
        header = parse_headers(lines[0])
        for each_line in lines[1:6]:
            value = parse_value(each_line)
            result.append(item_dict(header, value))
    return result
        
```
