**Q: (Optional) What is the overall life expectancy across in the world?

*Hint: You'll need to take a weighted average of life expectancy using populations as weights
*** To my best of knowledge, there is no direct function to calculate the weighted average for pandas as we have seen in numpy array. we therefore create a function to handle the problem for us

```python
def weighted_avg(countries_df, life_expectancy, population):
    expectancy = countries_df[life_expectancy]
    weights = countries_df[population]
    try:
        return (expectancy * weights).sum() / weights.sum()
    except ZeroDivisionError:
        return x.mean()
```

```python
overall_life_expectancy = weighted_avg(countries_df, "life_expectancy", "population")
```

```python
overall_life_expectancy
```

```python
# I'm using groupby() attribute to group the average according to continents.
# But for it to work, we must using the apply() attribute to apply grouping of the averages 
continents_life_expectancy = countries_df.groupby("continent").apply(weighted_avg, "life_expectancy", "population")
```

```python
continents_life_expectancy
```
